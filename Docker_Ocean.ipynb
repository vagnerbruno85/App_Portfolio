#Introdução ao Docker#




**Maquina virtual**

É uma representação de um ambiente físico(máquina/computador) por meio de software e pode ser hospedada dentro de outro sistema operacional. Uma máquina virtual possui sua propia arquitetura, que é criada virtualmente a partir do ambiente físico no qual essa máquina está hospedada.


**Containner**

Diferentes das maquinas virtuais os containners não virtualiza o ambiente completo, o que o torna mais leve e performatico. Em tese o containner trabalha apenas o sistema operacional, com foco na aplicação em questão e suas dependências.



**Docker**

o Docker é uma plataforma aberta que utiliza do conceito de containner para o desenvolvimento, implantação e execução de aplicações em ambientes isolados. A aplicação é empacotada se transformando em uma imagem docker.

**Usando o docker**

* docker run hello-world

**Listar imagens Docker**
* sudo docker images

**Mais opções sobre o comando**
* sudo docker images --help

**Baixar uma imagem**
* docker pull [nome da imagem]

**Criando o container a partir da imagem**
* docker run [nome da imagem]

**Inspecionando a Imagem**
* docker image inspect [nome da imagem]

**Exibindo containers criados**
* docker ps [nome da imagem]

**Exibindo containers ativos**
* docker ps -a [nome da imagem]

**Inicia o conatiner em modo interativo**
* docker run -it [nome da imagem]

**Reiniciar um processo parado**
* docker start -i CONTAINER_ID

**Enviando uma imagem para o DockerHub**

* ***Fazendo login na conta do DockerHub***

>* docker login

* ***Criando a imagem no formato para o upload***

>* docker commit container_id nomedousuario/repositório

>* docker commit container_id nomedousuario/repositório:tag


* ***Fazendo upload da imagem para o DockerHub***

>* docker push nomedousuario/repositório
